```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = "100"
lie_type = "attack_all_outrageous"
```

```python
statuses = {0: "no endpoint blacklisted", 1: "one endpoint blacklisted", 2: "both endpoints blacklisted"}
colors = {"no endpoint blacklisted": "green", 
          "one endpoint blacklisted": "blue", 
          "both endpoints blacklisted": "red"}
```

```python
data= pd.read_csv("data/percentage/test_"
                  +N+"_nodes_"+lie_type+".csv")[["lie_percentage","nb_strikes","blacklist_status", "threshold"]]
```

```python
data["status"] = data["blacklist_status"][data["blacklist_status"] > 0 ].map(statuses)
```

```python
data
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
grouped = data.groupby('status')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='lie_percentage', y='nb_strikes', label=key, color=colors[key])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.axhline(y=data["threshold"][0], xmin=0, xmax=1)

plt.show()
```

```python
fig.savefig('graphs/percentage/'+N+'_nodes_' + lie_type+".png")
```
